```python
## Lambda Functions in Python

"""
Lambda functions are small anonymous functions (functions without name) defined using the lambda keyword. They can have any number of arguments but only one expression (logic). They are commonly used for short operations or as argument to higher order functions.
Syntax:
lambda arguments: expression

"""
```

```python
def addition(a,b):
    return a+b
addition(2,3)
```

```python
addition = lambda a,b: a+b
print(type(addition))
print(addition(5,6))
```

```python
def even(num):
    if num % 2 == 0:
        return True
    return False

print(even(24))
```

```python
check_even = lambda num: num % 2 == 0
print(check_even(12))
```

```python
def addition(x,y,z):
    return x+y+z

addition(12,13,14)
```

```python
addition = lambda x,y,z : x+y+z
print(addition(12,13,14))
```

```python
## map() - applies a function to all items in an iterable; map has lazy loading means not intially intialized
numbers = [1,2,3,4,5,6]
def square(num):
    return num ** 2

print(list(map(lambda x : x ** 2,numbers)))
```
