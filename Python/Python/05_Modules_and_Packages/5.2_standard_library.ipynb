## Standard Library Overview

Python's Standard Library is a vast collection of modules and packages that come bundled with Python, providing a wide range of functionalities out of the box.

```python
import array
arr = array.array('i',[1,2,3,4])
print(arr)
```

```python
import math
print(math.sqrt(45))
print(math.pi)
print(math.pow(2,3))
```

```python
import random
random_number = random.randint(1,10)
print(random_number)
random_choice = random.choice(['apple','mango','cherry','litchi','strawberry'])
print(random_choice)
```

```python
### File and Directory Access

import os
print(os.getcwd())
# creating a folder in current working directory
# os.mkdir("test_dir")

file_path = os.path.join("test_dir","myfile.txt")
os.makedirs(os.path.dirname(file_path))

with open(file_path,'w') as file:
    file.write("File created.")
print(f"File '{file_path} created successfully.'")
```

```python
## high level operations on files and collection of files

import shutil
shutil.copyfile('source.txt', 'destination.txt')
```

```python
## Data Serialization
import json
data = {
    'name' :'Suhani',
    'age' : 21
}
print(type(data))
json_str = json.dumps(data, indent = 4)
print(json_str)
print(type(json_str))

## Converting JSON str back to dict
parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## csv (Comma Separated Values)
import csv
with open('example.csv',mode = 'w',newline = '') as file:
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Suhani',21])
    writer.writerow(['Frieren',48])

with open("example.csv", mode = 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
## datetime

from datetime import datetime,timedelta
now = datetime.now()
print(f"Current date and time: {now}")
yesterday = now - timedelta(days = 1)
print(f"Yesterday: {yesterday}")
```

```python
## time
import time
print(time.time())
# sleep for sometime
time.sleep(5)
print(time.time())
```

```python
## Regular expression
import re
pattern = r'\d+'
text = "There are 123 apples 456."
match = re.search(pattern,text)
print(match.group())

text2 = "This is a book"
pattern2 = r'This is'
match2 = re.search(pattern2,text2)
print(match2.group())
```
