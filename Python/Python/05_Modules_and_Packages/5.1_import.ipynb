## Importing Modules in Python: Modules and Packages

With the help of modules, I can directly use built-in functions, without writing the code from scratch.

In python, modules and packages help organize and reuse code.

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(25))
print(pi)
```

```python
# import numpy ## if the module is not installed we get "ModuleNotFoundError" so first we need to install the module in order to work with it.
## !pip install numpy 
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
my_array = np.array([1,2,3,4,5,6])
print(my_array)
```

```python
import numpy as np
from numpy import array

squares = array([1,4,9,16,25,36])
print(squares) # array
print([1,2,3,4,5]) # list
```

```python
# importing all the functions from a module
from math import *
print(sqrt(36))
print(pi)
```

```python
## Creating our own custom modules and packages
from package import mymath as math
print(math.addition(2,5))

from package.mymath import addition as add
print(add(5,6))
```
