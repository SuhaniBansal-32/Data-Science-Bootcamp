```python
## Basic Syntax Rules In Python
## Case sensitivity - Python is case sensitive
## Both these variables are different identifiers

name = "Suhani"
Name = "Bansal"
print(name)
print(Name)
```

```python
## Indentation
## Python uses indentation to define blocks of code. Consistent use of spaces (commonly 4) or a tab is required.

age = 32
if age > 30:
    print(age)
print(age) ## outside of the if clause
```

```python
## this is a single line comment
print("Hello world!")
```

```python
## Line Continuation
## Use a backslash(\) to continue to the next line

total = 1+2+3+4+5+\
4+5+6
print(total)
```

```python
## Multiple Statements on a single line
x = 5; y=10; z=x+y
print(z)
```

```python
## Understanding Semantics In Python
## Variable assignment
age = 32 ## age is an integer
name = "Suhani" ## name is a string
## Type inference : type define during runtime -> dynamic-typing
print(type(age))
print(type(name))
```

```python
## Type Inference
## Type determined at runtime : dynamic typing
variable = 10
print(type(variable))
variable = "Suhani"
print(type(variable))
```
