```python
## Declaring and assigning variables
age = 21
height = 5
name = "Suhani"
is_student = True

##printing the variables
print("age : ", age)
print("height : ", height)
print("name : ", name)
print("is_student : ", is_student)
```

```python
## Naming conventions
## variables names should be descriptive
## they must always with a letter or underscore
## It can contain letters, nubers or underscores in b/w
## variables names are case sensitive

## valid variable names
first_name = "Suhani"
last_name = "Bansal"
print(first_name, last_name, end = " ")
```

```python
## Invalid variable names
# 2age = 30
# first-name = "Suhani"
# @name = "Suhani"
```

```python
## case sensitivity
name = "Suhani"
Name = "Bansal"
print(name)
print(Name)
```

```python
## Understanding variable types
## python is dynamically typed, type of variable determined at runtime 
age = 25    #int
height = 5.1    #float
name = "Suhani" #string
is_student = True   #bool

print(type(age))
print(type(is_student))
```

```python
## Type conversion
age = 21
print(type(age))
age_str = str(age)
print(age_str)
print(type(age_str))
```

```python
## Dynamic Typing
## python allows the type of variable to change as the program executes
var = 10
print(var,type(var))
var = "Hello"
print(var,type(var))
var = 3.14
print(var,type(var))
```

```python

## creating a simple calculator
num1 = float(input("Enter first number : "))
num2 = float(input("Enter second number : "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum: ", sum)
print("Difference: ", difference)
print("Product: ", product)
print("Quotient: ", quotient)
```
