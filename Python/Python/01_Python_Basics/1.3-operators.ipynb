```python
## Comparison Operators
## == Equal to
a = 10
b = 10
result = (a==b)
print(result)

str1 = "Suhani"
str2 = "Suhani"
print(str1 == str2)

## Not Equal to !=
str1 != str2

## greater than >
num1 = 45
num2 = 55
num1 > num2 # False

## less than <
num1 < num2 # True

## greater than or equal to >=
number1 = 45
number2 = 45
number1 >= number2 # True

## less than or equal to
number1 <= number2 # True
```
