```python
## strings
## string is a data structure; collection of characters; each char of string can be iterated over

str = "Suhani Bansal"
for i in str:
    print(i)
```

```python
## while loop

## the while loop continues to execute as long as the condition is True

count = 0

while count < 5:
    print(count)
    count += 1
```

```python
count = 0
while count % 2 == 0:
    print(count)
    count += 1
```

```python
## Loop Control Statements

## break
## The break statement exits the loop prematurely.

## break statement

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continue
## The continue statement skips the current iteration and continues the next.

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
## pass
## The pass statement is a null operation; it does nothing. It acts as a placeholder for future code.

for i in range(5):
    if i == 3:
        pass # does nothing
    print(i)
```

```python
## Nested loops
## a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples - Calculate the sum of first N natual numbers using a while loop and for loop

## using while loop
num = int(input("Enter the number : "))
sum = 0
count = 1

while(count <= num):
    sum += count
    count += 1

print(sum)

## using for loop
sum = 0
for i in range(num+1):
    sum += i
print(sum)
```

```python
## Example - Prime numbers between 1 and 100
## Prime numbers are numbers that are divisible by 1 and the number itself
print("Prime numbers between 1 and 100 : ")
for num in range(1,101):
    if num > 1:
        is_prime = True
        for div in range(2,num):
            if num % div == 0:
                is_prime = False
                break
        if(is_prime):
            print(num,end = " ")
        
```

```python
## method - 2
for num in range(1,101):
    if num > 1:
        for div in range(2,num):
            if num % div == 0:
                break
        else:
            print(num)
```
