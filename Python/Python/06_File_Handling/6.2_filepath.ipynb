```python
## create a new directory
import os
# print(os.getcwd())
new_directory = "package"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' created.")
```

```python
## Listing Files and Directories
## . means current working location/ current working directory
items = os.listdir('.') # lists all the files and folders in current working directory
print(items)
```

```python
## Joining Path
dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(os.getcwd(),dir_name,file_name)
print(full_path)
```

```python
## checking if a path(file or folder) exists or not. It is checked in current working directory.
path = "example1.txt"
if os.path.exists(path):
    print(f"The path '{path}' exists.")
else:
    print(f"The path '{path}' does not exist.")
```

```python
# Checking if a Path is a File or Directory: The path will be checked as a file or folder if it exists in current working directory.
import os

path = "example1.txt"
if os.path.isfile(path):
    print(f"The path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path} is neither a file nor a directory.'")
```

```python
## relative path: path of a file or folder wrt another folder
## absolute path: complete path of a file or folder

## Getting the absolute path
relative_path = "example.txt" # means 'example.txt' is present in the current working directory
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
print(os.path.join(os.getcwd(),"package",relative_path))
```
