### Introduction to Dictionaries
Dictonaries are unordered collection of items.They store data in key-value pairs.Keys must be unique and immutable(e.g. strings,numbers,or tuples),while values can be of any type.

```python
## Creating Dictionaries
empty_dict = {}
print(empty_dict)
print(type(empty_dict))
```

```python
empty_dict = dict()
print(empty_dict)
```

```python
student = {"name":"Suhani","age":21,"country":"India"}
print(student)
```

```python
# Single key is always used
student = {"name":"Suhani","age":21,"name":"India"}
print(student)
```

```python
## Accessing Dictionary elements
student = {"name":"Suhani","age":21,"country":"India","grade":'A'}
print(student)
print(student["name"])
print(student["grade"])
print(student["age"])
print(student["country"])
# print(student["last_name"]) if the key is not present it throws KeyError but get() method returns None and avoid error if a certain key is not present

## Accessing using get() method: this method allows to give a default value if a certain key is not present
print(student.get("country"))
print(student.get("last_name"))
print(student.get("last_name","Not Available"))
```

```python
## Modifying Dictionary Elements
## Dictionary are mutable, so you can add, update or delete elements

print(student)
student["name"] = "Nikki" # updation
student["gender"] = "Female" # addition
print(student)
```

```python
del student["gender"] # delete key and value pair
print(student)
```

```python
## Dictionary methods
keys = student.keys() # get all the keys
print(keys)
print(type(keys))
values = student.values() # get all the values
print(values)
print(type(values))
items = student.items() # get all the key-value pairs
print(items)
print(type(items))
```

```python
## Shallow Copy : 2 diff dict exists, and change done in one dict is not reflected in another. They point to different reference
print(student)
student_copy = student # deep copy:once I assign one dict to another using '=', changes done in one also reflects in another. They point to same reference
print(student)
print(student_copy)

student["name"] = "Suhani"
print(student)
print(student_copy)

# shallow copy : using the copy() method. Different memory location is allocated, it takes the refrence to another memory location
student_copy1 = student.copy() 
print(student_copy1)
print(student)

student["name"] = "Nikki"
print(student_copy1)
print(student)
```

```python
### Iterating Over Dictionaries
## You can use loops to iterate over dictionaries, keys,values,items

# by default if we just iterate like for i in dictionary we get key

# Iterating over keys
for key in student.keys():
    print(key)
```

```python
# Iterating over values
for value in student.values():
    print(value)
```

```python
# Iterating over items(key-value pairs)
for key,value in student.items():
    print(f"{key}: {value}")
```

```python
### Nested Dictionaries
students ={
    "student1" : {"name":"Suhani","age":21},
    "student2" : {"name":"Yamini","age":2},
    "student3" : {"name":"Frieren","age":48},
}
print(students)
for student in students.values():
    print(student)
```

```python
## Accessing Nested Dictionaries elements
print(students["student1"]["name"])
print(students["student2"]["name"])
print(students["student3"]["name"])
```

```python
## Iterating over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key,value in student_info.items():
        print(f"{key}: {value}")
    print("-------------------")
```

```python
## Dictionary Comprehension
# Example : squares of specific numbers between 1-10
squares = {num : num ** 2 for num in range(1,11)}
print(squares)
```

```python
# Conditional Dictionary Comprehension
# Example : squares of only even numbers b/w 1-10
even_squares = {num:num*2 for num in range(11) if num % 2 == 0}
print(even_squares)
```

```python
## Practical Examples

## Use a dictionary to count the frequency of elements in list
# Method - 1
lst = [1,2,2,3,4,4,5,6]
frequency_dict = {el:lst.count(el) for el in lst}
print(frequency_dict)

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency = {}
# Method - 2
for number in numbers:
    if number not in frequency:
        frequency[number] = numbers.count(number)
print(frequency)

frequency2 = {}
# Method - 3
for number in  numbers:
    if number in frequency2:
        frequency2[number] += 1
    else:
        frequency2[number] = 1
print(frequency)

# Method - 4
frequency3 = {}
for number in numbers:
    frequency3[number] = frequency3.get(number,0)+1
print(frequency3)
```

```python
## Merge two dictionaries into one
dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}

merged_dict = {**dict1,**dict2}
print(merged_dict)
```
