### Sets
Sets are a built-in data type in Python used to store collections of unique items.They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements.Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference and symmetric difference.

```python
## Create a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(my_empty_set)
print(type(my_empty_set))
```

```python
number_set = set([1,2,3,4,5,5])
print(number_set)
# print(number_set[0]) TypeError: 'set' object is not subscriptable; sets are unordered
```

```python
## Basic Sets Operation
# adding and removing elements
my_set = {1,2,3,4,5}
print(my_set)
my_set.add(6)
print(my_set)
my_set.remove(2)
print(my_set)

# if an element is not present in a set and we try to remove it; an error is thrown
# my_set.remove(7)

# to prevent error even while removing an element that is not present in set we can use discard() method instead of remove() method

my_set.discard(7)
print(my_set)
```

```python
# set is FIFO
# pop method in set removes the first inserted element
print(my_set)
popped_el = my_set.pop() # pop() removes the first el of set and returns it
print(popped_el)
print(my_set)
```

```python
## clear all the elements
print(my_set)
my_set.clear()
print(my_set)
```

```python
## Set Membership Test: checks whether an el is a member of a set or not

my_set = {1,2,3,4,5}
print(3 in my_set)
print(7 in my_set)
print(7 not in my_set)
```

```python
## Mathematical Operations in Set:
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
## Union
union_set = set1.union(set2)
print(union_set)
## Intersection : returns the common elements from 2 sets
intersection_set = set1.intersection(set2)
print(intersection_set)

## doing sets intersection and updating first set to the intersection set
print(f"Original set1: {set1}")
set1.intersection_update(set2)
print(f"Set1 after intersection update: {set1}")
```

```python
## Difference Operation: displays all the els from first set that are not present in set2 (unique to set1)
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
difference_set = set1.difference(set2)
print(difference_set)

## removing common els from set2
difference_set = set2.difference(set1)
print(difference_set)
```

```python
## Symmetric difference : remove the common elements of the sets and only includes the unique elements from both the sets
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
symmetric_difference_set = set1.symmetric_difference(set2)
symmetric_difference_set = set2.symmetric_difference(set1)
print(symmetric_difference_set)
```

```python
# Sets Methods

set1 = {1,2,3}
set2 = {3,4,5}

# issubset
print(set2.issubset(set1))
# issuperset
print(set1.issuperset(set2))
```

```python
## Removing duplicates from a list
lst = [1,2,2,3,4,4,5]
print(f"List with duplicates: {lst}")
print(f"List without duplicates: {set(lst)}")
```

```python
### Example: Counting unique words in text
text = "In this tutorial we are discussing about sets"
words_list = text.split(" ")
## convert list of words to set to get unique words
unique_words = set(words_list)
print(unique_words)
print(len(unique_words))
```
