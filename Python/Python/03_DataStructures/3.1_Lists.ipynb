```python
## List Methods

cartoon_characters = ["Bheem","Doraemon","Hatori","Shinchan","Dora"]
print(cartoon_characters)

# a) append(el) : add an item to the end of list
cartoon_characters.append("Chiku")
print(cartoon_characters)

# b) insert(idx,el): to add an element at a specific index
cartoon_characters.insert(1,"Chutki")
print(cartoon_characters)

# c) remove(el): removes the first occurrence of element. It doesn't return anything
character = cartoon_characters.remove("Chiku")
print(character)
print(cartoon_characters)

# d) pop(): removes and return the last item of list
last_character = cartoon_characters.pop()
print(last_character)
print(cartoon_characters)

# e) index(el): returns the idx of specified el
idx_of_bheem = cartoon_characters.index("Bheem")
print(idx_of_bheem)
# It raises error if an element is not present in list

# f) count(el): count the total number of occurrences of an element in a list
cartoon_characters.append("Dora")
cartoon_characters.append("Dora")
cartoon_characters.append("Dora")
dora_count = cartoon_characters.count("Dora")
print(dora_count)

# g) sort(): It sorts the list in ascending order
cartoon_characters.sort()
print(cartoon_characters)

# f) reverse(): It reverse the original list
print(f"Original List: {cartoon_characters}")
cartoon_characters.reverse()
print(f"Reversed List: {cartoon_characters}")

# g) clear(): It empties the entire list/ it removes all items from the list
cartoon_characters.clear()
print(cartoon_characters)
```

```python
## List comprehension
# square of 1st 20 even numbers
squares = [x**2 for x in range(21) if x % 2 == 0]
print(squares)

# another example
odd_even_list = ['O' if x % 2 == 1 else 'E' for x in range(1,21)]
print(odd_even_list)
```

### List Comprehension
---
- Basic Syntax: [expression for item in iterable]
- with conditional logic:
    - syntax: [expression for item in iterable if condition]
    - nested list comprehension: 
    syntax: [expression for item1 in iterable1 for item2 in iterable2]

```python
# nested list comprehension
lst1 = [1,2,3,4]
lst2 = ['a','b','c','d']

# making pairs of els from lst1 and lst2
# pair = []
# for i in lst1:
#     for j in lst2:
#         pair.append((i,j))
# print(pair)

pair = [(i,j) for idx1,i in enumerate(lst1) for idx2,j in enumerate(lst2) if idx2 == idx1]
print(pair)
```

```python
## Using built-in functions with list comprehension
## List Comprehension with Function Calls
words = ["hello","world","python","list","comprehension"]
lengths = [len(word) for word in words]
print(lengths)


```
