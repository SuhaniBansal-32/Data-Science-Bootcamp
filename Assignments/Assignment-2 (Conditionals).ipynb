```python
## Solution - 15
## Printing Fibonacci Sequence

n = int(input("Enter the range, n for fibonacci numbers : "))
a,b = 0,1
for i in range(n):
    print(a, end = " ")
    a,b = b,a+b
```

```python
## Solution - 14
## Prime Number Check

user_num = int(input("Enter your number : "))
is_prime = True
if(user_num <= 1):
    is_prime = False

for div in range(2,user_num):
    if(user_num % div == 0):
        is_prime = False
        break
else:
    is_prime = 1

## else block with the loop executes if the loop is not terminated by the break statement

print(is_prime)
if(is_prime):
    print(f"{user_num} is a prime number.")
else:
    print(f"{user_num} is not a prime number.")
```
