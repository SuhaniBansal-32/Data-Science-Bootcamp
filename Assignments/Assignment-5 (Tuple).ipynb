```python
### Solution - 10
my_tuple = tuple(num for num in range(1,6)) # tuples are immutable  means they can't be modified
# converting tuple to list for modification
print(f"Original Tuple: {my_tuple}")
my_list = list(my_tuple)
my_list.append(6)
# converting back to tuple after modification is done
my_tuple = tuple(my_list)
print(f"Updated Tuple: {my_tuple}")
```

```python
### Solution - 11
my_str = "Hello! I am Suhani Bansal"
# making a tuple from the characters of my_str
my_tuple = tuple(char for char in my_str)
print(f"Tuple with string characters: {my_tuple}")
# converting my_tuple back to my_str
my_str = ("").join(my_tuple)
print(f"My String: {my_str}")
```

```python
### Solution - 15
numbers = (11,2,43,57,89,100,7)

def find_sum(tup):
    return sum(tup)

def find_min(tup):
    return min(tup)

def find_max(tup):
    return max(tup)

sum_of_tuple = find_sum(numbers)
min_in_tuple = find_min(numbers)
max_in_tuple = find_max(numbers)

print(f"Sum of tuple elements: {sum_of_tuple}")
print(f"Minimum element in tuple: {min_in_tuple}")
print(f"Maximum element in tuple: {max_in_tuple}")
```
