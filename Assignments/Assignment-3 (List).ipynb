```python
## Solution - 6 : sorted() function returns new list without modifying the original list
import random
random_numbers = [random.randint(1,20) for _ in range(15)]
print(f"Original List: {random_numbers}")

sorted_numbers = sorted(random_numbers)
print(f"Sorted in ascending order: {sorted_numbers}")

sorted_numbers_desc = sorted(random_numbers, reverse = True)
print(f"Sorted in descending order: {sorted_numbers_desc}")

unique_random_numbers = list(set(random_numbers))
print(f"List with duplicates removed: {unique_random_numbers}")
```

```python
## Solution - 15
def list_with_common_elements(lst1,lst2):
    intersected_lst = [num1 for num1 in lst1 for num2 in lst2 if num1 == num2]
    return intersected_lst

print(list_with_common_elements([1,2,3,4,5],[1,2,4]))
```

```python
## Solution - 13
def reverse_lst_func(lst):
    rev_lst = lst[::-1]
    return rev_lst
original_lst = [1,2,3,4,5]
print(f"Original list: {original_lst}")
print(f"Reversed list: {reverse_lst_func(original_lst)}")
```

```python
## Solution - 14

def rotate_list_func(n,lst):
    n = n % len(lst)
    if(n == 0):
        return lst
    right_half_els = lst[-n:]
    left_half_els = lst[:len(lst)-n]
    return (right_half_els + left_half_els)
lst = [1,2,3,4,5,6,7]
n = 7
print(f"original list: {lst}")
print(f"rotated list: {rotate_list_func(n,lst)}")
```

```python
## Solution - 10
def flatten_lst(nested_lst):
    lst = list(nested_lst[0])
    for i in range(1,len(nested_lst)):
        lst.extend(nested_lst[i])
    print(lst)


flatten_lst([[1,2],[3,4]])

# using list comprehension
nested_lst = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

def flatten_lst2(nested_lst):
    lst = [item for sublist in nested_lst for item in sublist]
    return lst

print(f"Original list: {nested_lst}")
flattened_lst = flatten_lst2(nested_lst)
print(f"Flattened list: {flattened_lst}")

```
