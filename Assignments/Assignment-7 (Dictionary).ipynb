```python
## Solution - 15
import json
book = {
    'title' : "Atomic Habits",
    'author' : "James Clear",
    'year' : 2001,
    'genre' : 'self-help'
}

# converting the dictionary to JSON string
json_book = json.dumps(book,indent = 4)
print(f"Book in JSON string {type(json_book)} format:\n {json_book}")
```

```python
## Solution - 14
import json
def count_characters(my_str):
    words = {}
    my_str = my_str.replace(" ","")
    for char in my_str:
        char = char.lower()
        if char in words:
            words[char] += 1
        else:
            words[char] = 1
    return json.dumps(words,indent = 4)

str_name = "Suhani Bansal"
words_count = count_characters(str_name)
print(f"Word character count in {str_name}:\n{words_count}")
```

```python
## Solution - 13
default_dict = {
    1 : [],
    2 : [],
    3 : [],
    4 : [],
    5 : []
}
s,e = 1,4
for value in default_dict.values():
    for i in range(s,e):
        value.append(i)
    s = e
    e += 3

print(f"Dictionary after updation: {(default_dict)}")
```

```python
## Solution - 12
squares_dict = {num : num ** 2 for num in range(1,6)}
swapped_dict = {}
for key,value in squares_dict.items():
    key,value = value,key
    swapped_dict[key] = value

print(f"Original dictionary:\n {squares_dict}")
print(f"Dictionary after swapping:\n {swapped_dict}")
```

```python
## Solution - 11
squares = {num : num**2 for num in range(1,11)}
# filtered dictionary where only even key is present
even_keys = list(filter(lambda key: key % 2 == 0, squares))
filtered_dict = {num : num ** 2 for num in even_keys}
print(f"Dcitionary with even keys only: {filtered_dict}")
```
